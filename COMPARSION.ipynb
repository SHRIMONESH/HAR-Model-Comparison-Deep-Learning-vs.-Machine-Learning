```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create DataFrame from the given data
data = {
    'Model': ['LSTM', 'CNN-LSTM', 'ML Model'],
    'Accuracy': [90.02, 91.08, 90.00],
    'Precision': [90.31, 91.04, 91.00],
    'Recall': [90.02, 91.08, 90.00],
    'F1-Score': [90.01, 91.01, 90.00]
}

df = pd.DataFrame(data)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Set width of bars
barWidth = 0.2

# Set positions of the bars on X axis
r1 = np.arange(len(df['Model']))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

# Create bars
ax.bar(r1, df['Accuracy'], width=barWidth, label='Accuracy', color='#4472C4')
ax.bar(r2, df['Precision'], width=barWidth, label='Precision', color='#ED7D31')
ax.bar(r3, df['Recall'], width=barWidth, label='Recall', color='#A5A5A5')
ax.bar(r4, df['F1-Score'], width=barWidth, label='F1-Score', color='#70AD47')

# Add labels and title
ax.set_xlabel('Models', fontweight='bold', fontsize=12)
ax.set_ylabel('Percentage (%)', fontweight='bold', fontsize=12)
ax.set_title('Model Performance Comparison', fontweight='bold', fontsize=14)
ax.set_xticks([r + barWidth*1.5 for r in range(len(df['Model']))])
ax.set_xticklabels(df['Model'])

# Add grid lines for better readability
ax.grid(True, linestyle='--', alpha=0.7, axis='y')

# Set y-axis to start at 85% for better visualization of differences
ax.set_ylim(85, 92)

# Add value labels on top of each bar
for i, bars in enumerate([r1, r2, r3, r4]):
    metric = ['Accuracy', 'Precision', 'Recall', 'F1-Score'][i]
    for j, bar in enumerate(bars):
        value = df[metric].iloc[j]
        ax.text(bar, value + 0.1, f'{value}%', ha='center', va='bottom', fontsize=9)

# Add legend
ax.legend(loc='lower right')

# Add a footnote for additional context
plt.figtext(0.5, 0.01, 'Note: Y-axis starts at 85% to better visualize the differences between models', 
            ha='center', fontsize=8, style='italic')

# Adjust layout
plt.tight_layout(pad=3.0)

# Show the plot
plt.show()
```
